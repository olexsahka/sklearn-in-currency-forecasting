```python
import pandas
usd=pandas.read_excel("usd.xlsx")
```

```python
past=7
future=7
```

```python
money=usd.curs
```

```python
start=past
end=len(money)-future
```

```python
table=[]
for i in range(start,end):
    s=money[i-past:i+future]
    table.append(list(s))
```

```python
past_columns=[]
for i in range (past):
    past_columns.append(f"past_{i}")
```

```python
future_columns=[]
for i in range (future):
    future_columns.append(f"future_{i}")
```

```python
df=pandas.DataFrame(table,columns=(past_columns+future_columns))
```

```python
x=df[past_columns][:-10]
y=df[future_columns][:-10]
x_test=df[past_columns][-10:]
y_test=df[future_columns][-10:]
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.neural_network import MLPRegressor
```

```python
LR=LinearRegression()
#в модель mlp добавил 1 скртый слойу, также изменил максимальное колличетсво иттераций , поставил скорость обучения для обновления веса
kn=KNeighborsRegressor(n_neighbors=1)
ml=MLPRegressor(hidden_layer_sizes=(100,100,100,100,100),max_iter=10000,random_state=42,learning_rate_init=0.00001)
ml.fit(x,y)
```

```python
# 1model MLP
prediction=ml.predict(x_test.head(1))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
for i in range (10):
    prediction=ml.predict([x_test.iloc[i]])
    err=mean_absolute_error(prediction[0],y_test.iloc[i])
    print (f"day#{i},error={err}")
    plt.plot(prediction[0],label="prediction")
    plt.plot(y_test.iloc[i],label="real data")
    plt.legend()
    plt.show()
```
